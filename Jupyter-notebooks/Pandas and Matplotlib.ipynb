# Linear regression using scipy package

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import matplotlib

matplotlib.rc('xtick', labelsize=14)     
matplotlib.rc('ytick', labelsize=14)
font = {'size'   : 14}
matplotlib.rc('font', **font)

#------------------For Sample5-------------------
print('---Put a file as logLx(x),logLr(y) coloumns---')
file5= 'Sm5.txt'
x5,y5=np.loadtxt(file5,skiprows = 1,delimiter="\t", dtype='float', comments="#", unpack=True)

# syntax slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
slope, intercept, r_value, p_value, std_err = stats.linregress(x5, y5)
print("slope: %f    intercept: %f" % (slope, intercept))
print("R-squared: %f" % r_value**2)
print("standard deviation=",std_err)
#plt.style.use('seaborn')
plt.plot(x5[0:107], y5[0:107], 'o', label='SMBH',color='blue', )
plt.plot(x5[108:157], y5[108:157], '*', label='SBH',color='red')
plt.plot(x5, intercept + slope*x5, 'r', label='fitted line',color='black')

plt.xlabel("0.64 log L$_X$ + 0.76 log M")
plt.ylabel("log L$_R$ (erg.s$^{-1}$)")

userinput5 = 'FP-BH.png'
plt.legend()
plt.savefig(userinput5,dpi=300,bbox_inches='tight',pad_inches=0.2)
plt.show()
```
